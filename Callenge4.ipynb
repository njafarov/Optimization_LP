```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory
```

```python
from pyomo.core.base import initializer
model = pyo.ConcreteModel()

model.i = pyo.Set(initialize = ['Product High', 'Product Supreme'])

model.A = pyo.Param(model.i, initialize={'Product High':3, 'Product Supreme':2})
A = model.A
model.B = pyo.Param(model.i, initialize={'Product High':1, 'Product Supreme':3})
B = model.B

model.x = pyo.Var(model.i, domain=pyo.NonNegativeReals)
x = model.x 
def objective_function(model):
  return 80*x['Product High']+200*x['Product Supreme'] 
model.Obj = pyo.Objective(rule = objective_function, sense=pyo.maximize)

def cons1(model,i):
  return sum(A[i]*x[i] for i in model.i)<=300
model.Cons1 = pyo.Constraint(rule=cons1)

def cons2(model,i):
  return sum(B[i]*x[i] for i in model.i)<=280
model.Cons2 = pyo.Constraint(rule=cons2)

Solver = SolverFactory('glpk')
results = Solver.solve(model)

print(results)
print("Objective: ", model.Obj())

for i in model.i:
  print(i,"= ", x[i]())
```
